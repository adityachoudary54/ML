```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dfx=pd.read_csv("xdata.csv")
dfy=pd.read_csv("ydata.csv")
```

```python
X=dfx.values
Y=dfy.values
X=X[:,1:]
Y=Y[:,1]
print(X.shape)
print(Y.shape)
```

```python
query_x=np.array([1,1])
plt.scatter(X[:,0],X[:,1],c=Y)
plt.scatter(query_x[0],query_x[1],c='red')
plt.show()
```

```python
b=np.array([3,4,5])
a=np.zeros(3,)
print(np.sqrt(abs(np.sum((a-b)**2))))
```

```python
def dist(a,b):
    return np.sqrt(abs(np.sum((a-b)**2)))
def knn(X,Y,query_x,k=5):
    vals=[]
    m=X.shape[0]
    for i in range(m):
        d=dist(query_x,X[i])
        vals.append((d,Y[i]))
    vals=sorted(vals)
    #Nearest K points
    vals=vals[:k]
    vals=np.array(vals)
    new_vals=np.unique(vals[:,1],return_counts=True)
    index=new_vals[1].argmax()
    prediction=new_vals[0][index]
    return prediction
```

```python
print(knn(X,Y,query_x))
```

```python
#Calculating accuracy
cnt=0.0
m=X.shape[0]
for i in range(m):
    pred=knn(X,Y,X[i])
    if(pred==Y[i]):
        cnt+=1
print(cnt/399)
```
