```python
from sklearn.datasets import load_digits
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
dfx=pd.read_csv('linearX.csv')
dfy=pd.read_csv('linearY.csv')
x=dfx.values
y=dfy.values

#Normalise
x2=(x-x.mean())/x.std()
plt.scatter(x,y,label="Raw data")
plt.scatter(x2,y,c='orange',label="Normalised")
plt.legend()
plt.show()
```
