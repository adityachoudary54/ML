# Histogram

```python
# Generate a normal distribution(1-D)
u=0
sigma=1
# such a distribution is called standard normal distrbution
vals=np.random.randn(1000)
print(vals.shape)
plt.hist(vals,50)
plt.show()
```

```python
# for any random u and sigma proceed like this
u=5
sigma = 0.2
vals=u + sigma*np.random.randn(200)
print(vals.shape)
#print(vals)
plt.hist(vals,10)
plt.show()
```

```python
vals=np.round(vals)
z=np.unique(vals,return_counts=True)
print(z)
```

```python
x=vals
y=np.zeros(x.shape)
plt.scatter(x,y)
plt.show()
```

# Multivariate Normal distribution

```python
#Bivariate
mean=np.array([0.0,0.0])
cov=np.array([[1,0],[0,1]])
dist=np.random.multivariate_normal(mean,cov,500)
print(dist.shape)
plt.figure(figsize=(6,6))
plt.scatter(dist[:,0],dist[:,1])
mean2=np.array([5.0,6.0])
cov2=np.array([[1.3,-0.2],[-0.2,1.1]])
dist2=np.random.multivariate_normal(mean2,cov2,500)
print(dist2.shape)
plt.scatter(dist2[:,0],dist2[:,1])
plt.show()
```
