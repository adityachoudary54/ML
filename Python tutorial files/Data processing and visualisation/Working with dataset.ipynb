```python
title=ds.get("movie_title")
print(type(title))
```

```python
title=np.array(title)
freq_title={}
for ix in title:
    len_title=len(ix)
    if freq_title.get(len_title)==None:
        cnt=0
    else:
        cnt=freq_title.get(len_title)
    freq_title[len_title]=cnt+1
plt.plot(freq_title.keys(),freq_title.values())
```

```python
plt.style.available
```

```python
# code as in video
freq_title={}
for ix in title:
    leng=len(ix)
    try:
        freq_title[leng]+=1
    except:
        freq_title[leng]=1
plt.title('seaborn-dark')
plt.style.use('bmh')
plt.scatter(freq_title.keys(),freq_title.values(),label="Title lengths")
plt.legend()
plt.show()
```
