```python
#Create new numpy arrays
arr_zero=np.zeros((5,5))
print(arr_zero)
```

```python
for ix in range(5):
    for ij in range(5):
        arr_zero[ix][ij]=ix+ij
print(arr_zero)
```

```python
arr3=np.random.randint(low=-5,high=10,size=(4,6))
print(arr3)
```

```python
#shape of array
print(arr3.shape)
print(arr3.dtype)
```

```python
#Reshape, Transpose, Concatenate numpy arrays
arr4=arr3.reshape((6,-1))
print(arr4)
print(arr4.shape)
print(arr4.T)#Transpose
arr5=np.array([[1,2],[4,5],[3,5],[4,5]])
print(arr5.shape)
arr6=arr5.reshape((8,))
arr7=arr5.reshape((8,1))
print(arr6)
print(arr7)
```

# Matplotlib

```python
#Plot simple ranges
a=np.array(range(0,100,2))
b=np.array(range(0,100,2))
plt.figure(0)#giving id
plt.title("Line")
plt.grid("on")#optional
plt.xlim(0,100)
plt.ylim(0,100)
plt.plot(a,b,'r--')# hyphens are used to create dotted line
plt.figure(1)
plt.plot(a,2*a,'g')
plt.show()
```

```python
#Plot simple ranges
plt.style.use('seaborn-dark')
a=np.array(range(0,100,2))
b=np.array(range(0,100,2))
plt.figure(0)#giving id
plt.title("Line")
plt.grid("on")#optional
plt.xlim(0,100)
plt.ylim(0,100)
plt.plot(a,b,'r--',label="2018")# hyphens are used to create dotted line
plt.plot(a,2*a,'g',label="2017")
plt.xlabel("year")
plt.ylabel("price")
plt.legend()# used to print labels
plt.show()
```

```python
#Plot simple ranges
a=np.array(range(0,100,2))
b=np.array(range(0,100,2))
plt.figure(0)#giving id
plt.title("Line")
plt.grid("on")#optional
plt.xlim(0,100)
plt.ylim(0,100)
plt.plot(a,b,'r--',label="2018")# hyphens are used to create dotted line
#plt.figure(1)
plt.scatter(a,2*a,c='g',label="2017",marker="*")
plt.legend()
plt.show()
```

```python
#Plot sine wave, cosine wave and some lines
x=np.linspace(-2*np.pi,2*np.pi,num=100)
plt.figure(0,figsize=(20,5))
plt.plot(x,np.sin(x),'r')
plt.plot(x,np.cos(x),'g')

plt.show()
```

```python
#Plot sine wave, cosine wave and some lines
x=np.linspace(-2*np.pi,2*np.pi,num=100)
plt.figure(0,figsize=(20,5))
plt.plot(np.sin(x),'b')
plt.plot(np.cos(x),'y')
plt.show()
```

# Pandas

```python
df=pd.read_csv("movie_metadata.csv")
#last 10 rows
df.tail(10)
#first 5 rows
print(df.head(5))
df[21:31]
```

```python
df.info()
```

```python
# get the values as numpy array and check shape
values=df.values
print(values.shape)
print(type(values))
```

```python
#For printing columns
df.columns
```

```python
#Print a few movie titles
titles=df.get("movie_title")
lens=[]
for ix in titles.values:
    lens.append(len(ix))
lens.sort()
#print(titles)
#print(lens)
```

```python
#create and plot frequency of each title length
freq={}
for ix in lens:
    cur=freq.get(ix,0)
    freq[ix]=cur+1
```

```python
plt.style.use('seaborn-dark')
plt.plot(freq.keys(),freq.values())
```
