```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_x=pd.read_csv("./Datasets/Locally Weighted Regression/weightedX.csv")
data_y=pd.read_csv("./Datasets/Locally Weighted Regression/weightedY.csv")
print(data_x.shape)
print(data_y.shape)
```

```python
X=data_x.values
Y=data_y.values
```

```python
plt.style.use("seaborn")
u=X.mean()
std=X.std()
X=(X-u)/std
plt.scatter(X,Y)
```

```python
ones=np.ones((X.shape[0],1))
X=np.hstack((ones,X))
```

```python
X.shape
```

```python
def weight(x,query_x,t):
    w=2.718**(-((x-query_x).sum()/2*t**2)**2)
    return w
def closed_form_sol_weighted(X,Y,query_x):
    X_trans=np.transpose(X)
    Y_trans=np.transpose(Y)
    a=[weight(X[i],query_x,2) for i in range(X.shape[0])]
    w=np.diag(a)
    theta=(np.linalg.inv(np.dot(X_trans,w).dot(X))).dot(np.dot(X_trans,w).dot(Y))
    return theta
def hypothesis(theta,x):
    return np.dot(theta,x)
def score(X,Y,Y_pred):
    ss_tot=0
    ss_res=0
    y_mean=np.mean(Y)
    for i in range(Y.shape[0]):
        ss_res=ss_res+(Y[i]-Y_pred[i])**2
        ss_tot=ss_tot+(Y[i]-y_mean)**2
    return 1-ss_res/ss_tot
```

```python
res=[]
for i in range(X.shape[0]):
    a=closed_form_sol_weighted(X,Y,X[i])
    a=a.transpose()
    res.append(hypothesis(a,X[i,:]))
```

```python
plt.scatter(X[:,1],res)
```

```python
print("The validation score is {0:.3f}".format(float(score(X,Y,res))))
```
